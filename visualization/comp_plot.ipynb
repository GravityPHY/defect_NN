```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_lc(title,data,metric_type):
    fig, ax = plt.subplots()
    labels = range(1000,(len(data)+1)*1000,1000)
    lc_all_train=[]
    lc_all_test=[]
    x=np.arange(len(labels))
    width=0.35
    if metric_type=="MAE":
        ax.set_ylabel("MAE(eV)")
        for pt in data:
        #all_train_err=[]
        #all_test_err=[]
        #all_train_err.append(pt["trainsoap_mean MAE"])
        #all_test_err.append(pt["testsoap_mean MAE"])
            lc_all_test.append(list(pt.values())[1])
            lc_all_train.append(list(pt.values())[2])
        #lc_all_train.append(val for key, val in pt.items() if "MAE" and "test" in key)
        #lc_all_test.append(val for key, val in pt.items() if "MAE" and "train" in key)
    elif metric_type=='RMSE':
        ax.set_ylabel("RMSE(eV)")
        for pt in data:
            lc_all_test.append(list(pt.values())[3])
            lc_all_train.append(list(pt.values())[4])
    elif metric_type=="MSE":
        ax.set_ylabel("MSE(eV^2)")
        for pt in data:
            lc_all_test.append(list(pt.values())[3]**2)
            lc_all_train.append(list(pt.values())[4]**2)
    elif not metric_type:
        raise NotImplementedError
    ax.set_title(title)
    rects1=ax.bar(x-width/2,lc_all_test,width,label='Test')
    rects2=ax.bar(x+width/2,lc_all_train,width,label='Train')

    ax.set_title(title)
    ax.set_xticks(x,labels)
    ax.set_xlabel("Training size")
    ax.legend()

    ax.bar_label(rects1,padding=3)
    ax.bar_label(rects2,padding=3)

    fig.tight_layout()
    plt.show()
    #return ax.figure.savefig(title+'.png')
```
